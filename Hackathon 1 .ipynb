```python
topFiveSales = df[['price', 'quantity', 'Sales']].sort_values(by='Sales', ascending=False).head(5)
print(topFiveSales)
```

```python
topFiveSales = df.sort_values(by='Sales', ascending=False).head(5)
```

```python
plt.figure(figsize=(8,4))
ax = sns.barplot(data=topFiveSales, x='product', y='Sales', palette='coolwarm')
plt.title('Top 5 Best Selling Products')
plt.xlabel('Product')
plt.ylabel('Sales')
for p in ax.patches:
    value = p.get_height()                  
    x = p.get_x() + p.get_width() / 2       
    y = value                               

    ax.text(x, y + 5000,                    
            f'{int(value)}',                
            ha='center', fontsize=12, fontweight='bold')
plt.show()
```

## Observations:
The chart indicates that “Headphones” and “Bags” dominate sales, each reaching close to 400,000 units. In contrast, “Mobiles” and “Shoes” show lower figures, averaging around 320,000. Overall, within the top five products, “Headphones” remain slightly ahead, while “Mobiles” and “Shoes” fall in the mid-tier range. The pattern also suggests that buyers tend to favor electronic items and accessories more than footwear.

```python
payment_counts = df['payment_method'].value_counts()
print(payment_counts)
```

```python
plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title("Payment Method Distribution")
plt.show()
```

## Observation:
The pie chart illustrates that customers rely on a variety of payment methods, with usage spread out quite evenly. JazzCash emerges as the top choice, with Credit Card and Easypaisa close behind. COD also accounts for a significant portion, indicating that a good number of customers still favor paying at the time of delivery. Debit Card appears to be the least popular option, showing minimal preference compared to digital wallets and credit cards. Overall, digital payment methods remain the most dominant in the distribution.

```python
monthwise_sales = df.groupby('Month')['Sales'].sum()
print(monthwise_sales)
```

```python
plt.figure(figsize=(10,6))
plt.plot(monthwise_sales.index, monthwise_sales.values, marker='s', linestyle='-', color='green')
plt.title('Monthwise Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45) 
plt.show()
```

## Observation:
The line graph indicates that monthly sales vary sharply over the year. March shows the peak sales, with January, April, and May also performing well. On the other hand, July, August, and September reflect much weaker sales figures. Overall, the pattern reveals recurring highs and lows rather than a steady rise or decline, suggesting that sales are likely affected by seasonal trends or other external influences.

```python
df['City'] = df['City'].replace({
    'New york': 'New York',
    'LA': 'Los Angeles',
    'San fran': 'San Francisco'
})
print(df['City'].unique())
```

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values()
city_sales
```

```python
city_names = ["Karachi", "Lahore", "Islamabad", "Faisalabad", "Quetta", "Peshawar"]

plt.figure(figsize=(10,6))
ax = sns.barplot(x=city_sales.index, y=city_sales.values, palette='Set2')
plt.title('City-wise Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')

for i, value in enumerate(city_sales.values):
    ax.text(i,max(city_sales.values) + (max(city_sales.values) * 0.03),  
        city_names[i],
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold',
        rotation=0) 
    
plt.ylim(0, max(city_sales.values) * 1.15)   # space upar add kar diya
plt.show()
```

## Observations:
The bar graph presents a comparison of total sales across six cities. City 2 records the highest sales, making it the top-performing location. City 5 also achieves strong results but falls short of City 2. Cities 1, 4, and 6 register the lowest sales, reflecting weaker market activity in these regions. City 3 lies in the mid-range with moderate sales. Overall, the graph highlights an uneven distribution, with some cities outperforming others significantly.

```python
monthly_category_sales = df.pivot_table(index='Month', columns='category', values='Sales', aggfunc='sum')
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_category_sales = monthly_category_sales.reindex(month_order)
```

```python
months = monthly_category_sales.index
categories = monthly_category_sales.columns
num_months = len(months)
num_categories = len(categories)

bar_width = 0.8 / num_categories  
positions = range(num_months)

colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1', '#F4A261']

plt.figure(figsize=(12,6))

for i, category in enumerate(categories):
    plt.bar(
        [p + i*bar_width for p in positions],
        monthly_category_sales[category],
        width=bar_width,
        label=category,
        color=colors[i % len(colors)]   
    )

plt.xticks([p + bar_width*(num_categories-1)/2 for p in positions], months)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales by Category')
plt.legend()
plt.show()

```

## Observations:
The bar chart illustrates monthly sales for Accessories, Electronics, and Fashion. Accessories see strong sales in January and March but decline in the middle of the year. Electronics maintain a relatively steady performance, with spikes in March, May, and October. Fashion sales vary, performing well in February, May, and December, while dipping during mid-year months. Overall, March and May emerge as high-sales months, whereas June through September record lower sales across all categories.

## Overall Observation:
The analysis shows that a few products generate the majority of sales, with Electronics and Fashion leading the list. Credit Card is the preferred payment method, while Easypaisa and JazzCash are less commonly used. Karachi and Lahore stand out as the top-performing cities in terms of sales. December records the highest sales, indicating a seasonal pattern. These insights can guide inventory management, marketing strategies, and payment options to increase revenue and improve customer satisfaction.
