#### Check Duplicate Records

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("cars.csv")

duplicates = df[df.duplicated()]
duplicates
```

#### Drop Duplicate Records

```python
df = df.drop_duplicates()
df
```

#### Check missing or null values

```python
df.isnull().sum()
```

#### Drop missing or null values

```python
df = df.dropna()
df
```

#### Show top 10 average MSRP (Price) per car brand (rounded figure)?

```python
df.columns
```

```python
avg_price = df.groupby('Make')['MSRP'].mean().round()
top10_avg_price = avg_price.sort_values(ascending=False).head(10)
top10_avg_price
```

#### Show the average MSRP (price) per years (rounded figure)?

```python
avg_price_per_year = df.groupby('Year')['MSRP'].mean().round()
avg_price_per_year = avg_price_per_year.sort_index()
avg_price_per_year
```

#### Show average Engine HP (power) rounded figure and MPG (efficiency) rounded figure per year?

Hint: Average of (city mpg + highway MPG)

```python

df['Average MPG'] = (df['city mpg'] + df['highway MPG']) / 2
avg_stats_per_year = df.groupby('Year').agg({
    'Engine HP': 'mean',
    'Average MPG': 'mean'
}).round()  # Round the figures
avg_stats_per_year = avg_stats_per_year.sort_index()
avg_stats_per_year
```

#### What percentage of cars are Compact, Midsize, or Large?

```python

size_counts = df['Vehicle Size'].value_counts()

size_percentage = (size_counts / df.shape[0] * 100).round(2)

size_percentage_filtered = size_percentage[['Compact', 'Midsize', 'Large']]
size_percentage_filtered
```

#### Which vehicle styles are the most sold (most common) in the dataset?

```python
style_counts = df['Vehicle Style'].value_counts()
style_counts
```

#### Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
filtered_df = df[(df['MSRP'] >= 40000) & (df['Popularity'] >= 1000)]
brand_counts = filtered_df['Make'].value_counts()
brand_counts
```

#### Check if there are outliers in Engine HP show boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.boxplot(df['Engine HP'].dropna(), vert=True)
plt.title("Boxplot of Engine HP")
plt.ylabel("Engine HP")
plt.show()
```

#### Check if there are outliers in Engine Cylinder show boxplot

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
plt.boxplot(df['Engine Cylinders'].dropna(), vert=True)
plt.title("Boxplot of Engine Cylinders")
plt.ylabel("Engine Cylinders")
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Engine Cylinders'])
plt.title("Boxplot of Engine Cylinders")
plt.ylabel("Engine Cylinders")
plt.show()
```

#### Find upper bound, lower bound, number of outliers in Engine HP

```python
Q1 = df['Engine HP'].quantile(0.25)
Q3 = df['Engine HP'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Engine HP'] < lower_bound) | (df['Engine HP'] > upper_bound)]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of outliers in Engine HP:", outliers.shape[0])
```

#### Fill missing values in Number of Doors column with median value then check if values are filled

```python

median_doors = df['Number of Doors'].median()
df['Number of Doors'].fillna(median_doors, inplace=True)

missing_after_fill = df['Number of Doors'].isnull().sum()

print("Median value used to fill missing Number of Doors:", median_doors)
print("Number of missing values after filling:", missing_after_fill)
```
