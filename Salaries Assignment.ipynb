```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Salaries.csv')
df
```

Drop rows where EmployeeName and JobTitle are 'Not provided'

```python
df = df[~((df['EmployeeName'] == 'Not provided') & (df['JobTitle'] == 'Not provided'))]
df
```

Replace 'Not Provided' with 0 in BasePay OvertimePay	OtherPay	Benefits

```python

cols_to_replace = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
df[cols_to_replace] = df[cols_to_replace].replace('Not Provided', 0)
df
```

Change BasePay	OvertimePay	OtherPay	Benefits type to float

```python
cols_to_float = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
df[cols_to_float] = df[cols_to_float].apply(pd.to_numeric, errors='coerce').fillna(0)
print(df[cols_to_float].dtypes)
```

Display all employees name and basepay columns who have a BasePay greater than 200,000.

```python
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce').fillna(0)
high_basepay = df[df['BasePay'] > 200000][['EmployeeName', 'BasePay']]
print(high_basepay)
```

Find all employees whose JobTitle contains the word “Manager” (case = False)

```python
managers = df[df['JobTitle'].str.contains('Manager', case=False, na=False)]
managers
```

Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.

```python

df_selected = df[['EmployeeName', 'JobTitle', 'TotalPay']]
df_selected
```

Find count of employees who have zero OvertimePay.

```python
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'], errors='coerce').fillna(0)
zero_overtime_count = (df['OvertimePay'] == 0).sum()
zero_overtime_count
```

Calculate the TotalPay for each JobTitle where totalpay > 300000

```python
totalpay_by_job = df.groupby('JobTitle')['TotalPay'].sum()
high_totalpay_jobs = totalpay_by_job[totalpay_by_job > 300000]
high_totalpay_jobs
```

Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order

```python
avg_totalpay_by_job = df.groupby('JobTitle')['TotalPayBenefits'].mean()
top5_jobs = avg_totalpay_by_job.sort_values(ascending=False).head(5)
top5_jobs
```

Replace all missing Benefits values with 0

```python

df['Benefits'] = df['Benefits'].fillna(0)
df['Benefits'].head()
```

Identify the top employee name by TotalPay in each Year. Show year, employeename and totalpay dataframe

```python

top_employee_idx = df.groupby('Year')['TotalPay'].idxmax()
top_employees = df.loc[top_employee_idx, ['Year', 'EmployeeName', 'TotalPay']]
top_employees = top_employees.reset_index(drop=True)
top_employees
```

Find the number of employees per JobTitle.

```python

employees_per_job = df['JobTitle'].value_counts()
employees_per_job
```
