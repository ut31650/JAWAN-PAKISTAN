```python
import pandas as pd
import sqlite3 as sql
conn = sql.connect('database.sqlite')
conn
```

```python
pd.read_sql_query('select * from salaries', conn)
```

```python
def sql_query(u):
    return pd.read_sql_query(u,conn)
```

#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

```python
u = '''select JobTitle, AVG(TotalPayBenefits) AS Avg_TotalPayBenefits
    From salaries where JobTitle not in ('Not Provided', 'Not provided')
    group by JobTitle order by Avg_TotalPayBenefits desc'''
sql_query(u)
```

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

```python
u = '''select JobTitle, OvertimePay
    from Salaries where overtimepay > 50000'''


sql_query(u)
```

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

```python
u = '''select distinct(employeename) from salaries 
    where employeename != 'Not provided' '''

sql_query(u)
```

#### **4.Show The Top 10 highest Paid Employees**

```python
u= ''' select employeename , totalpay from salaries 
   order by totalpay desc limit 10'''
sql_query(u)
```

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
u= ''' select employeename , TotalPay Benefits from salaries 
   order by totalpay desc '''
sql_query(u)
```
